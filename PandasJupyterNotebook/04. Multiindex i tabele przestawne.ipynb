## Swaplevel
zmiana kolejności indexów w multiindexach
<br> Standardowo najpierw resetujemy index, a następnie jeszcze raz przydzielamy indexy poprzez set_index()

```python
incidents = pd.read_csv("./data/Canadian Railway Crossing Incidents.csv")
```

```python
incidents.set_index(["Region", "EventType"], inplace=True)
```

```python
incidents.sort_index(inplace=True)
```

### .swaplevel(i=, j=)
i= numer pierwszego indexu, który chcemy zamienić 
<br> j = numer drugiego indexu, który chcemy zamienić 
<br> nie podając 'i' i 'j' zamienione zostaną dwa najbardziej wewnętrzne indexy

```python
incidents.swaplevel().head(10)
```

```python
incidents.swaplevel().sort_index().head(10)
```

Żeby zapisać DF, trzeba przypisać do zmiennej

```python
incidents = incidents.swaplevel().sort_index()
```

swaplevel na kolumnach (axis='columns')

```python
incidents.swaplevel(axis=1).head(10)       # nie zadziała, bo jest tylko jeden poziom kolumn
```

### .unstack()
Domyślnie z najbardziej wewnętrznego poziomu indexu tworzy poziom kolumn

```python
stackedIncidents.unstack().head(10) 
```

```python
stackedIncidents.unstack(0).head(10)            # pierwszy index zostanie przeniesiony do kolumn 
```

```python
stackedIncidents.unstack(level="Region").head(10)   
```

```python
stackedIncidents.unstack(level=["Region", "EventType"]).head(10)   
```
