# Oczyszczanie danych

### .dropna()
Funckja dropna() w dataFrame domyślnie usunie wszystkie wiersze, w których brakuje chociaż jednej wartości 

### .dropna(how="all")
w tym przypadku wiersze zostaną usunięte tylko wtedy kiedy wszystkie wartości będą NaN. Domyślnie = "any"

### .dropna(subset=["TotalFat], how="any")
subset = wpisujemy listę kolumn, które mają być analizowane. Usuwa wiersz jeśli w wymienionych kolumnach znajdzie wartość NaN

### .dropna(axis="rows")
domyślnie axis = 0, czyli rows. Wtedy usuwane są wiersze 

### .dropna(axis="columns")
axis = 1  == axis ="columns" - wtedy usuwane są kolumny, w których będą jakieś wartości NaN

### .dropna(inplace="True")
usuwa na stałe w danym dataFrame bez przypisywania zmiennej. Domyślnie jest False

## Metoda .fillna()
Metoda, która pozwala uzupelniać brakujące dane

### fillna(value=None, method=None, axis=None, inplace=False, limit=None)
takie są wartości domyślne metody .fillna()
<br> value= wartość jaką chcemy wpisać zamiast NaN (np. 0 lub jakiś napis)
<br> 

Żeby określiś lepiej wartości jakie chcemy mieć w konkretnych kolumnach, możemy stworzyć słownik, gdzie kluczami będą nazwy kolumn, a wartościami wartości, które chcemy mieć w konkretnych kolumnach zamiast NaN. Taki słonik później przypisujemy do wartości value w metodzie .fillna()

```python
replaceDefinitionTable = {"Category":"UNKNOW", "TotalFat":0,"Saturated Fat":0}
```

```python
replaceDefinitionTable
```

```python
dataFrame.fillna(value=replaceDefinitionTable).head()
```

Zamieni wartości NaN w podanych kolumnach na podane wartości. 

### .fillna() tylko w jednej kolumnie

```python
dataFrame["Category"].fillna(value="UNKNOW", inplace=True)
```

### przypisywanie średniej zamiast wartości NaN
wyliczamy średnią w nowej zmiennej --> zmienną wstawiamy jako value w metodzie .fillna()

```python
avgTotalFat = dataFrame["TotalFat"].mean()
```

```python
dataFrame["TotalFat"].fillna(avgTotalFat, inplace=True)
```

### .fillna(method=)
method="ffill" - uzupełni wartości NaN wartością z wiersza powyżej. Wczęsniej warto posortować dataFrame po innej kolumnie, dzięki której ta kolumna z brakowymi wartościami będzie najbardziej zbliżona do wartości, która tam powinna być.
<br> method="bfill" uzupełni dane kolejną wartością ? nie potwierdzone

```python
dataFrame["Saturated Fat"].fillna(method="ffill", inplace=True)
```

```python
mcd = pd.read_csv("./data/mcdonalds.csv", usecols=["Item","Category", "Serving Size", "Calories", "TotalFat"])
```

```python
mcd.head()
```

gdy filtrujemy to cały argument musi znaleźćsię w (). trzymujemy wartość boolean (True or False)

```python
(mcd['Calories'] >= 400).head(10)
```

```python
(mcd['TotalFat']<=20).head(10)
```

Żeby wykorzystać filtrowanie w naszym data frame i otrzymać normalne wartości zamiast boolean musimy przypisać filtrowanie do zmiennej.

```python
caloriesGreaterEqual400 = mcd['Calories'] >= 400
```

Tutaj bez ()
<br> a następnie użyć utworzoną zmienną jako argument do całego data frame

```python
mcd[caloriesGreaterEqual400].head()
```

W taki sposób wyświetliliśmy 5 pierwszych wyrażeń, które spełniają kryteria filtrowania

```python
totalFatLessEqual20 = mcd['TotalFat']<= 20
```

```python
mcd[totalFatLessEqual20].head()
```

```python
isBreakfast = mcd['Category'] == 'Breakfast'
```

#### Łączenie filtrów 

```python
mcd[isBreakfast & caloriesGreaterEqual400].head()
```

Otrzymaliśmy dataFrame, z elementami, które są śniadaniami oraz mają kaloryczność większą niż lub równą 400

```python
mcd[isBreakfast & caloriesGreaterEqual400 & totalFatLessEqual20].head()
```

#### Znak negacji (odwrócenie kryterium fltrowania) "~"

```python
mcd[~ isBreakfast].head()
```

Wyświetliło wszystko to co nie jest śniadaniem

### Operatory filtrowania
& - i
<br> | - lub 
<br> & wykonywane jest w pierwszej kolejności, później "|". Kolejność warunków możemy zmienić używając (), tak jak w matematyce

```python
mcd[caloriesGreaterEqual400 | totalFatLessEqual20 & isBreakfast].head(100)
```

Wyświetliliśmy wszystkie dania, które są śniadaniem i ich totalFat jest mniejszy lub równy 20, a następnie wypisaliśmy wszystkie elementy, które mają kaloryczność większe lub równe 400 ( nie musiały być śniadaniem)

```python
mcd[(caloriesGreaterEqual400 | totalFatLessEqual20) & isBreakfast].head(100)
```

Tutaj natomiast dzęki nawiasowi wybraliśmy wszystkie produkty, które mają kaloryczność większą lub równą 400 oraz te które mają totalFat <= 20 i później sprawdziliśmy, czy te produkty są też śniadaniem. Dlatego wszystkie wyniki są śniadaniami.

# Filtrowanie .where i .query

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
mcd = pd.read_csv("./data/mcdonalds.csv", usecols=["Item","Category", "Serving Size", "Calories", "TotalFat"])
```

```python
ha400CalOrMore = mcd['Calories'] >= 400
```

```python
mcd.where(ha400CalOrMore).head(10)
```

Filtrowanie wcześniejszym sposobem tworzyło nowy dataFrame tylko z wartościami spełniającymi kryteria.
<br> W przypadku .where wyświetlany jest cały dataFrame, aczkolwiek w mejsca, gdzie nie były spełnione warunki wpisane jest NaN. Żeby się pozbyć wartości NaN z dataFrame użyjemy funkcji dropna()

```python
mcd.where(ha400CalOrMore).dropna().head()
```

### .query()
w () wpisujemy wyrażenia (kryteria), którymi chcemy filtrować tak jak w jezyku SQL.
<br> nazwy kolumny po której filtrujemy nie bierzemy w ()

```python
mcd.query('Category == "Desserts"').head()
```

"in" oznacza to samo co: ('Category == "Desserts" | Category == "Beverages"')

```python
mcd.query('Category in ["Desserts", "Beverages"]').head(20)
```

Możemy również od razu wpisywać więcej warunków.

```python
mcd.query('Category == "Desserts" and Calories < 200').head()
```

# Duplikaty
Jeżeli w dataFrame znajdują się powtarzające się dane, a my chcemy pracować tylko z unikalnymi danymi to je przetwarzamy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
color = ['green', 'green', 'blue', 'blue', 'red', 'red', 'white']
size = ['S','S','S','M','M','M','L']
gender = ['M', 'W', 'M', 'W', 'M', 'W', 'U']
clothes = pd.DataFrame({'color':color, 'size':size, 'gender':gender})
clothes
```

```python
clothes['color'].is_unique
```

#### .unique()
zwraca w formie listy wszystkie różne wartości występujące w danej kolumnie. Jeżeli mamy 4 razy 'red' i 3 razy 'blue' wyświetli raz red i raz blue

```python
clothes['color'].unique()
```

#### .nunique()
obliczy ilość unikalnych obiektów. Omija wartości NaN

```python
clothes['color'].nunique()
```

Jeśli chcemy policzyć również wartości NaN to używamy:

```python
len(clothes['color'].unique())
```

#### .duplicated()
wykrywa, czy istnieją wartości, które się powielają. Domyślnie - .duplicated(keep='first')

```python
clothes['color'].duplicated(keep='first')
```

.duplicated(keep='first') - zwraca False w miejscu, gdzie dana wartość występuje pierwszy raz. Tam gdzie wartość się duplikuje zwraca wartość True.

```python
clothes['color'].duplicated(keep='last')
```

(keep='last') - True przy wszystkich wynikach, które się powtarzają, poza ostatnią. Ostatnia wartość przyjmuje wartość False, tak samo wartości, które się nie powtarzaja 

```python
clothes['color'].duplicated(keep=False)
```

True przy wszytkich wartościach, które się powtarzają. False przy wartościach, które się nie powatrzają

```python
uniqueValues = clothes['color'].duplicated(keep='first')
clothes[uniqueValues]    # wyświetli powtórzenia, lub clothes[~uniqueValues] wyświetli wszystkie pierwsze wystąpienia wartości
```

#### .drop_duplicates(subset=['color'], keep='first')

```python
clothes.drop_duplicates(subset=['color'], keep='first')
```

zwrócone zostały wszystkie pierwsze wystąpienia

```python
clothes.drop_duplicates(['color', 'size'])
```

zwrócone zostały wszystkie elementy, gdzie nie powtarzały się elementy z obydwu kolumn (wiersz 1 jest taki jak 0, a wiesz 5 taki jak 4 - jeśli chodzi o kolumny color i size)
