### wykorzystanie utworzonej funkcji 
### .apply()

```python
def getComment(row):                        # wykonuję funkcję dla każdego wiersza. 
    reporterType = row['Reporter Type']
    cost = float(row['Total Operating Expenses'].replace("$", ""))
    
    if(cost > 200000):
        comment = "CLASS A"
    else:
        comment = "CLASS B"
    return (reporterType + "/" + comment)
```

```python
frame.apply(getComment, axis=1).head()
```

```python
fortune[["City", "State"]] = fortune['Hqlocation'].str.strip().str.split(",", expand=True)
```

```python
fortune.head()
```

```python
def BuildShortcut(row):
    industry = row["Industry"]
    result =""
    for i in industry.split(" "):
        result += i[0]
    
    return result
```

```python
BuildShortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune["IndustryShort"] = fortune.apply(BuildShortcut, axis=1)
```

## Przygotowanie danych po imporcie

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv("./data/PublicTransitExpenses.csv", low_memory=False)
```

```python
frame.head()
```

```python
frame.info(memory_usage='deep')
```

Zdarza się tak, że Python nie jest w stanie określic jaki jest typ danej kolumny i przypisuje jej typ object, przez co wykorzystywane jest więcej pamięci. W takich przypadkach do załadowania df potrzebujemy parametru low_memory=False.
<br> Żeby zmniejszyć ilość wykorzystywanej pamięci zaciągajmy tylko te kolumny, których potrzebujemy do analizy

```python
frame = pd.read_csv('./data/PublicTransitExpenses.csv', usecols=["Agency", "Reporter Type", "Organization Type", 
                                                                 "Rail (Y/N)", "Fixed Route (Y/N)", "Service Costs", 
                                                                'Tires and Tubes', "Total Operating Expenses", 
                                                                 "Service Area Population"], low_memory=False)
```

```python
frame.head()
```

```python
frame.info(memory_usage='deep')
```

Samo zaczytanie tylko wybranych kolumn zmniejszyło wykorzystywaną pamięc z 30MB do 9MB
<br> Jeśli chcemy to możemy zmienić nazwy kolumn, żeby się prościej pracowało

```python
newColumnNames = {"Agency"                  :"Agency",
                  "Reporter Type"           :"ReporterType",
                  "Organization Type"       :"OrgType",
                  "Rail (Y/N)"              :"IsRail",
                  "Fixed Route (Y/N)"       :"IsFixedRoute",
                  "Service Costs"           :"ServiceCost",
                  'Tires and Tubes'         :"TiresTubesCost",
                  "Total Operating Expenses":"TotalExpenses",
                  "Service Area Population" :"Population"}
```

```python
frame.rename(columns=newColumnNames, inplace=True)
```

```python
frame.head()
```

Sprawdzamy, czy warto konwertować kolumnę do typu category

```python
frame['ReporterType'].nunique()
```

```python
frame['ReporterType'].count()
```

Kolumna ReporterType posiada 17844 elementów, które są uzupełnione 4 różnymi wartościami - Idealna do konwersji na typ category

```python
frame['ReporterType'].value_counts()
```

```python
frame['ReporterType'] = frame['ReporterType'].astype('category')
```

```python
frame.info(memory_usage='deep')
```

Zmniejszyliśmy wykorzystaną pamięć o kolejny MB

```python
frame['Agency'] = frame['Agency'].astype('category')
```

```python
frame['OrgType'] = frame['OrgType'].astype('category')
```

```python
frame.info(memory_usage='deep')
```

Już spadło do tylko 5MB
<br> Następnie sprawdzamy kolumnę Population, która powinna być Int, a jest Float(dlatego, że przyjmuje wartości NaN)
<br> Więc zmieniamy NaN na 0 i konwertujemy kolumne na Int

```python
frame['Population'] = frame['Population'].astype('int')
```

```python
frame[frame['Population']>0].head()
```
